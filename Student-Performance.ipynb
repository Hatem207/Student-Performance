# Summary

This analysis explores the key factors influencing studentsâ€™ final academic performance (G3) using demographic, academic, and behavioral data.

The findings indicate that **prior academic performance** is the strongest predictor of final outcomes. In particular, second period grades (G2) show a very strong positive correlation with final grades, highlighting the importance of continuous academic monitoring throughout the academic year.

Additionally, **study time** demonstrates a clear positive relationship with final performance, suggesting that structured and consistent study habits significantly improve academic results. Conversely, **past academic failures** and **high absenteeism** show strong negative relationships with final grades, identifying these variables as major risk indicators.

Behavioral factors such as **weekday alcohol consumption** also display a negative association with performance, although their impact is weaker compared to core academic indicators. This suggests that while lifestyle habits influence outcomes, academic history and attendance remain more critical determinants.

Overall, the analysis reveals that student performance is primarily driven by:

* Prior academic achievement
* Study effort
* Attendance consistency
* Academic history (failures)

These findings emphasize the importance of early identification of at-risk students, continuous performance tracking, and targeted academic support programs to improve overall educational outcomes.


# Recommendations
Schools should focus on early identification of low-performing students, strengthen academic support systems, encourage structured study habits, reduce risky behaviors, and enhance parental involvement through data-driven strategies.
